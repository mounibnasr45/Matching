```python
import pandas as pd
import random

# List of possible hard skills and soft skills
hard_skills = [
    "Python", "Java", "SQL", "Machine Learning", "Data Science", "JavaScript", "C++", "HTML", "CSS", 
    "React", "Node.js", "Spring", "REST APIs", "Django", "Data Analysis", "Data Visualization", "Deep Learning",
    "TensorFlow", "Keras", "Pandas", "NumPy", "AWS", "GCP", "Azure", "Docker", "Kubernetes", "Git"
]

soft_skills = [
    "Communication", "Teamwork", "Problem Solving", "Adaptability", "Leadership", "Creativity", 
    "Time Management", "Collaboration", "Critical Thinking", "Conflict Resolution", "Emotional Intelligence",
    "Public Speaking", "Decision Making", "Negotiation", "Stress Management"
]

# Generate 200 dummy profiles
profiles = []
for i in range(1, 201):
    # Random selection of 7 hard skills and 3 soft skills
    selected_hard_skills = random.sample(hard_skills, 7)
    selected_soft_skills = random.sample(soft_skills, 3)
    all_skills = selected_hard_skills + selected_soft_skills
    profile = {
        "id": i,
        "nom": f"User{i}",
        "competences": ", ".join(all_skills)
    }
    profiles.append(profile)

# Create a DataFrame
profiles_df = pd.DataFrame(profiles)

# Save to CSV
profiles_df.to_csv('profiles.csv', index=False)

# Check the first few rows of the generated data
profiles_df.head()
```
